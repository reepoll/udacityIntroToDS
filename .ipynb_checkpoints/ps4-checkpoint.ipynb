```python
%matplotlib inline
import pandas as pd
import numpy as np
from ggplot import *
```

```python
filename = 'hr_year.csv'
path = './data/'
```

```python
data = pd.read_csv(path + filename)
```

```python
data.head()
```

```python
ggplot(data, aes(x='yearID', y='HR')) +\
    geom_point(color='coral') + geom_line(color='blue') +\
    ggtitle('Total HR by Year') + xlab('HR')
```
